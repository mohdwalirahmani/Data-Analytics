DATA TRANSFORMATION (Feature Engineering)

```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
df2 = df.copy()
```

apply() allows us to apply a custom function to rows, columns, or elements of Pandas data structures.<br>
apply is Higher Order Function

```python
# apply()
df2["tax"] = df["income"].apply(lambda x : "20%" if x>=50_000 else "10%")
df2
```

=> map() applies a function to each element of an iterable or Pandas Series and returns transformed values.

```python
# map()
gender_map = {
    "Male" : "M",
    "Female" : "F",
    "Unknown" : "U"
}
df2["gender"] = df2["gender"].map(gender_map)
df2
```

When we run this code second time, it gives NaN. When map() is applied multiple times, already mapped values do not exist as keys in the dictionary, resulting in NaN. This is why replace() is safer for repeated execution.

| Feature   | `map()`                | `apply()`          |
| --------- | ---------------------- | ------------------ |
| Works on  | Series / Iterables     | Series / DataFrame |
| Operation | Element-wise           | Row / Column wise  |
| Speed     | Faster                 | Slower             |
| Use case  | Simple transformations | Complex logic      |


```python
# assign ()
df2 = df2.assign(new_income = df2["income"] * 1.1)    # inc income by 10%
df2
```

assign() is used to create or modify columns in a DataFrame and returns a new DataFrame, enabling clean method chaining.

```python
# replace(old, new)
df2["country"] = df2["country"].replace("USA","United States")
df2
```
