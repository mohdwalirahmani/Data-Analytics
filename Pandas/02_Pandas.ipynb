```python
df = pd.read_csv("raw_data.csv")
```

```python
# Data Cleaning - Handling Missing Values
df.isnull()
df.isnull().sum()      # this will tell kitne missing values h harr column m

df.dropna()       # drop row with missing value
df.dropna(axis = 1)      # drop col with missing value

df.fillna(0)   
# age_mean = df["age"].mean()
# df["age"].fillna(age_mean)

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
cleaned_data
```
