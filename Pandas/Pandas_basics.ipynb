Series : 1D labelled array <br>
DataFrame : 2D labelled array

```python
import pandas as pd
```

Series (Properties)
1. Homogenous
2. Vectorized Operations
3. Handle Missing values with NaN
4. Mutable values, immutable size

```python
s = pd.Series([1,2,3,4,5])
print(s)
print(type(s))

#index
print(s[2])
```

```python
s = pd.Series([56,54,23,21], index = ["Ram", "Mohan", "Wali", "kk"])      # labels
print(s)
print(s["Ram"])
```

```python
s1 = pd.Series([1,2,3,4,5,6])
s2 = pd.Series([10,20,30,40,50,60])

print(s1+s2)

s1[0] = 69
print(s1)
change = s1.drop(0)
print(change)
```

```python
# DataFrame
info = {
    "Name" : ["Aman", "Eve", "Alice"],
    "Age" : [23,56,65],
    "CGPA" : [7.2,6.4,9.2]
}
df = pd.DataFrame(info)
print(df, type(df))
print(df.index)
print(df.columns)
```

```python
# Creating df using list of lists
info = pd.DataFrame([["Adam",23], ["Bob",56], ["Alice", 65]], columns = ["Name", "Age"])
print(df)
```

```python
# creating df using np.array
import numpy as np
np_arr = np.array([[1,2,3],[4,5,6]])
df = pd.DataFrame(np_arr, columns = ["A","B","C"])
print(df)
```
