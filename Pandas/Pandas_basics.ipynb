Series : 1D labelled array <br>
DataFrame : 2D labelled array

```python
import pandas as pd
```

```python
s = pd.Series([1,2,3,4,5])
print(s)
print(type(s))

#index
print(s[2])
```

```python
s = pd.Series([56,54,23,21], index = ["Ram", "Mohan", "Wali", "kk"])      # labels
print(s)
print(s["Ram"])
```

```python
s1 = pd.Series([1,2,3,4,5,6])
s2 = pd.Series([10,20,30,40,50,60])

print(s1+s2)

s1[0] = 69
print(s1)
change = s1.drop(0)
print(change)
```
