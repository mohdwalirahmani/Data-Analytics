```python
import matplotlib.pyplot as plt
```

```python
# Bar Graphs
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo",
    "12 Years a Slave",
    "Birdman",
    "Spotlight"
]
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
oscar_revenue = [1005, 170, 427, 133, 232, 187, 103, 98]

plt.bar(years, oscar_revenue)
plt.title("Revenue in each year for Oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

# plt.text(years[0], oscar_revenue[0] + 10, str(oscar_revenue[0]), ha = "center")
for i in range (len(years)):
    plt.text(years[i], oscar_revenue[i] + 10, str(oscar_revenue[i]), ha = "center")

plt.ylim(0, max(oscar_revenue) + 100)           # it is used to inc the height of the graph
```

plt.text(): is used to write text directly on a plot — like labels, values, or notes inside the graph. <br>
plt.text(x, y, "your text")

```python
import numpy as np
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo",
    "12 Years a Slave",
    "Birdman",
    "Spotlight"
]
oscar_revenue = [1005, 170, 427, 133, 232, 187, 103, 98]
Non_Oscar_Movies = [
    "Avengers: Endgame",
    "Titanic",
    "Avatar",
    "Jurassic World",
    "The Lion King",
    "Frozen",
    "Inception",
    "Interstellar"
]
non_oscar_revenue = [980, 210, 390, 160, 260, 200, 120, 110]
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]

x = np.arange(len(years))
width = 0.4

plt.bar(x - width/2, oscar_revenue, width, label = "Oscar Movies")           # left shifting
plt.bar(x + width/2, non_oscar_revenue, width, label = "Non-Oscar Movies")       # write shifting

plt.title("Oscar vs Non-Oscar Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
plt.xticks(x ,years)             # plt.xticks(positions, labels)
```

```python
# horizontal graph
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo",
    "12 Years a Slave",
    "Birdman",
    "Spotlight"
]
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
oscar_revenue = [1005, 170, 427, 133, 232, 187, 103, 98]

plt.barh(oscar_movies, oscar_revenue)

plt.title("Revenue in each year for Oscar movies")
plt.xlabel("Revenue (in $M)")
plt.ylabel("Movies")
```
