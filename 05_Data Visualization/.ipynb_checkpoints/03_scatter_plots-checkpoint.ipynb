```python
import matplotlib.pyplot as plt
```

```python
# scatter plot
ppl = ["Person K", "Person L", "Person M", "Person N", "Person O",
          "Person P", "Person Q", "Person R", "Person S", "Person T"]

age = [24, 27, 31, 36, 42, 47, 52, 56, 61, 68]
bp = [112, 118, 121, 126, 129, 134, 138, 129, 147, 152]

colors = ["red" if x > 135 else "green" for x in bp]

plt.scatter(age, bp, color = colors)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

# plt.colormaps()
```

```python
# colormaps
plt.scatter(age,bp, cmap = "plasma", c=bp)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.colorbar()

# plt.annotate("Person S", xy = (61,147))        # annotate label the point
for i in range(len(ppl)):
    plt.annotate(ppl[i], xy = (age[i], bp[i]), xytext = (age[i]+1, bp[i]))
```

```python
# Multiple Plots (data from gpt)

cities = [
    "City A", "City B", "City C", "City D", "City E",
    "City F", "City G", "City H", "City I", "City J",
    "City K", "City L", "City M", "City N", "City O"
]

# Winter: Temperature (°C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7, 3, 6, 1, 4, 8, 9, 2, 5, 7, 3]
winter_humidity = [80, 75, 65, 85, 70, 78, 72, 88, 76, 68, 66, 82, 74, 71, 79]

# Summer: Temperature (°C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27, 32, 29, 34, 26, 31, 33, 24, 28, 30, 27]
summer_humidity = [60, 50, 55, 45, 65, 48, 53, 42, 67, 49, 46, 62, 54, 51, 64]

plt.scatter(winter_temp, winter_humidity, label = "Winter")
plt.scatter(summer_temp, summer_humidity, label = "Summer")

plt.title("Summer vs Winter Data")
plt.xlabel("Temp")
plt.ylabel("Humidity")
plt.legend()
```

```python
# pie chart

weather_types = ["Sunny", "Rainy", "Cloudy", "Snowy", "Stormy"]
days_count = [120, 90, 70, 40, 45]

explode = [0,0,0,0,0.2]      # for highlighting a data

plt.pie(days_count, labels = weather_types, autopct = "%1.1f%%",
        wedgeprops = {
            "edgecolor" : "white",
            "linewidth" : 1,
            "linestyle" : "--"
        },
        explode = explode,
        shadow = True
        
       )
# autopct shows % in chart

plt.title("Days in each Weather Types")
# plt.style.available
```
