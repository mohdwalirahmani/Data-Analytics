```python
# Real data from gpt
# oscar winning movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo",
    "12 Years a Slave",
    "Birdman",
    "Spotlight"
]

years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
oscar_revenue = [1005, 170, 427, 133, 232, 187, 103, 98]  # in $M

#plot
plt.plot(years, oscar_revenue)
plt.title("Oscar Winning Movies")
plt.xlabel("Years")
plt.ylabel("Oscar Revenue (in $M)")
```

```python
# Multiple datasets on line plot
# Popular Non-Oscar Movies
Non_Oscar_Movies = [
    "Avengers: Endgame",
    "Titanic",
    "Avatar",
    "Jurassic World",
    "The Lion King",
    "Frozen",
    "Inception",
    "Interstellar"
]
non_oscar_revenue = [2797, 2187, 2923, 1671, 1663, 1450, 836, 677]  # in $M

# plt.plot(years, oscar_revenue, label = "Oscar")
# format strings
# plt.plot(years, oscar_revenue, color="red", marker="o", linestyle="--", label = "Oscar")
plt.plot(years, oscar_revenue, "ro--", label = "Oscar")                 # we can also write like this
plt.plot(years, non_oscar_revenue, label = "Non-Oscar")

plt.title("Oscar vs Non-Oscar Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()                           # it is used for showing plot labels
```

```python
# Styling
plt.plot(years, oscar_revenue, label = "Oscar")
plt.plot(years, non_oscar_revenue, label = "Non-Oscar")

plt.title("Oscar vs Non-Oscar Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()

plt.style.available                 # it gives us all the possible styles
# plt.style.use("dark_background")
plt.grid()

'''
# xkcd is used for making cartoon type graph
with plt.xkcd():
    plt.plot(years, oscar_revenue, label = "Oscar")
    plt.plot(years, non_oscar_revenue, label = "Non-Oscar")
    
    plt.title("Oscar vs Non-Oscar Revenue")
    plt.xlabel("Years")
    plt.ylabel("Revenue (in $M)")
    plt.legend()
    
    plt.style.available
    plt.grid()
    plt.savefig("cartoon_plot.png")
    '''
```

## Plot Methods
1. plt.plot(x, y) -          Line plot
2. plt.bar(x, height) -      Vertical bar chart
3. plt.barh(y, width) -      Horizontal bar chart
4. plt.scatter(x, y) -       Scatter plot
5. plt.pie(data) -           Pie chart
6. plt.hist(data, bins) -    Histogram
7.plt.boxplot(data) -        Box plot
