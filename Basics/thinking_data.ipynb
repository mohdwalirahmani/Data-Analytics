```python
import json
```

```python
# Load the data
def load_dataset(filename):
    with open (filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_dataset("feedback_data.json")
print(data)
print(type(data))
```

```python
# CLean and Structure the Data
def clean_data(data):
    text_to_num = {
        "one": 1, 
        "two": 2, 
        "three": 3, 
        "four": 4
    }
    cleaned_data = []
    unique_users = set()          # set store only unique values
    
    for user in data:
        
        # Cleaning Ratings
        raw_rating = str(user.get("rating", "")).strip().lower()
        
        if raw_rating in text_to_num:
            user["rating"] = text_to_num[raw_rating]
        elif raw_rating.isdigit():
            user["rating"] = int(raw_rating)
        else:
            user["rating"] = None

        # Handling Missing Values
        user["age"] = user.get("age", None)       # if age exists return age else return none

        
        # Deduplication
        name = user.get("name", "").strip()
        if name in unique_users:
            continue
        
        unique_users.add(name)
        user["name"] = name
        cleaned_data.append(user)
        
    return cleaned_data
```

```python
clean_data(data)
```

```python
data = clean_data(data)
```

```python
# Getting insights from data

def get_insights(data):

    #avg rating
    total_rating = 0
    count = 0
    
    for user in data:
        if user["rating"] is not None:
            total_rating += float(user["rating"])
            count += 1
            
    print (f"Avg rating = {total_rating / count}")

    # percentage of users with poor rating
    poor_rating = 0
    for user in data:
        if user["rating"] is not None:
            if float(user["rating"]) < 3:
                poor_rating += 1
        
    print(f"Poor rating = {(poor_rating / count) *100}")
```

```python
get_insights(data)
```
