```python
import json
```

```python
# Load the data
def load_dataset(filename):
    with open (filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_dataset("feedback_data.json")
print(data)
print(type(data))
```

```python
# CLean and Structure the Data
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4}\
    
    for user in data:
        raw_rating = user["rating"]
        if (raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating
        print(raw_rating)
```

```python
clean_data(data)
```
